```python
import cv2
from ultralytics import solutions
# torch is already imported in cell 2, no need to import again

cap = cv2.VideoCapture("./media_files/People walking/computer_vision_object_and_detection_tracking_people_walking_video_20250730_164537_36.mp4")
# cap = cv2.VideoCapture("./media_files/istockphoto-1138941116-640_adpp_is.mp4")
assert cap.isOpened(), "Error reading video file"
w, h, fps = (int(cap.get(x)) for x in (cv2.CAP_PROP_FRAME_WIDTH, cv2.CAP_PROP_FRAME_HEIGHT, cv2.CAP_PROP_FPS))

# Define region points
region_points = [(150, 150), (1130, 150), (1130, 570), (150, 570)]

# Video writer
video_writer = cv2.VideoWriter("object_counting_output.avi", cv2.VideoWriter_fourcc(*"mp4v"), fps, (w, h))

# Init trackzone (object tracking in zones, not complete frame)
trackzone = solutions.TrackZone(
    show=True,  # display the output
    region=region_points,  # pass region points
    model="yolo11m.pt",
    conf=0.5,  # confidence threshold
    # iou=0.99,  # IoU threshold
    verbose=True,  # print additional information
    device="cuda:0",  # use GPU if available
)

# Process video
while cap.isOpened():
    success, im0 = cap.read()
    if not success:
        print("Video frame is empty or video processing has been successfully completed.")
        break
    results = trackzone(im0)
    video_writer.write(results.plot_im)

cap.release()
video_writer.release()
cv2.destroyAllWindows()
```

# Object Counting

```python
import cv2

from ultralytics import solutions


def count_objects_in_region(video_path, output_video_path, model_path):
    """Count objects in a specific region within a video."""
    cap = cv2.VideoCapture(video_path)
    assert cap.isOpened(), "Error reading video file"
    w, h, fps = (int(cap.get(x)) for x in (cv2.CAP_PROP_FRAME_WIDTH, cv2.CAP_PROP_FRAME_HEIGHT, cv2.CAP_PROP_FPS))
    video_writer = cv2.VideoWriter(output_video_path, cv2.VideoWriter_fourcc(*"mp4v"), fps, (w, h))

    region_points = [(20, 400), (1080, 400), (1080, 360), (20, 360)]
    counter = solutions.ObjectCounter(show=True, region=region_points, model=model_path)

    while cap.isOpened():
        success, im0 = cap.read()
        if not success:
            print("Video frame is empty or processing is complete.")
            break
        results = counter(im0)
        video_writer.write(results.plot_im)

    cap.release()
    video_writer.release()
    cv2.destroyAllWindows()


count_objects_in_region("path/to/video.mp4", "output_video.avi", "yolo11n.pt")
```

## Promt: rename video files dynamically with preset string from a folder using python on jupyter notebook code
### Deepseek understanding: Python script for dynamically renaming video files in a folder with a preset string pattern using Jupyter Notebook:

```python
import os
from datetime import datetime

def rename_videos(folder_path, preset_string, extension='.mp4'):
    """
    Renames all video files in a folder with a preset string + timestamp.
    
    Args:
        folder_path (str): Path to the folder containing videos.
        preset_string (str): Base name for files (e.g., "Security_Footage").
        extension (str): File extension to filter (default: '.mp4').
    """
    if not os.path.exists(folder_path):
        print(f"Error: Folder '{folder_path}' does not exist.")
        return
    
    files = [f for f in os.listdir(folder_path) if f.endswith(extension)]
    if not files:
        print(f"No {extension} files found in '{folder_path}'.")
        return
    
    print(f"Found {len(files)} video files to rename:")
    
    for idx, old_name in enumerate(files, 1):
        # Generate new name: preset_string + timestamp + index + extension
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        new_name = f"{preset_string}_{timestamp}_{idx}{extension}"
        
        old_path = os.path.join(folder_path, old_name)
        new_path = os.path.join(folder_path, new_name)
        
        os.rename(old_path, new_path)
        print(f"Renamed: '{old_name}' → '{new_name}'")
    
    print("Renaming complete!")

# Example Usage
# folder_path = "./media_files/People walking"  # Replace with your folder path
# preset_string = "CV_face_detection_and_recognition_people_walking"  # Customize this

# folder_path = "./media_files/People walking"  # Replace with your folder path
# preset_string = "CV_object_detection_traking_road_side"  # Customize this

folder_path = "./media_files/Apple counting"  # Replace with your folder path
preset_string = "CV_object_detection_and_recognition_apple_counting"  # Customize this



rename_videos(folder_path, preset_string)
```

## Promt: regenerate above-given funcion for all kind of video and image files extension. add file type as preset string. inside media_files folder, folder path slug or name also add to preset string
### Deepseek understanding: an enhanced Python function that dynamically renames all video and image files in a folder, incorporating the folder name into the preset string and categorizing by file type:


```python
import os
from datetime import datetime

def rename_media_files(folder_path, preset_string, verbose=True):
    """
    Renames all video and image files in a folder with a structured name:
    {preset_string}_{file_type}_{timestamp}_{index}.{ext}
    
    Args:
        folder_path (str): Path to the folder containing files.
        preset_string (str): Base name for files (e.g., "Event").
        verbose (bool): Print detailed logs if True.
    """
    # Supported video and image extensions
    VIDEO_EXTENSIONS = ('.mp4', '.mov', '.avi', '.mkv', '.flv', '.webm')
    IMAGE_EXTENSIONS = ('.jpg', '.jpeg', '.png', '.gif', '.bmp', '.tiff')
    ALL_EXTENSIONS = VIDEO_EXTENSIONS + IMAGE_EXTENSIONS
    
    if not os.path.exists(folder_path):
        print(f"Error: Folder '{folder_path}' does not exist.")
        return
    
    files = [f for f in os.listdir(folder_path) if f.lower().endswith(ALL_EXTENSIONS)]
    if not files:
        print(f"No video/image files found in '{folder_path}'.")
        return
    
    if verbose:
        print(f"Found {len(files)} files to rename:")
    
    for idx, old_name in enumerate(files, 1):
        # Extract file extension and determine type
        ext = os.path.splitext(old_name)[1].lower()
        file_type = "VIDEO" if ext in VIDEO_EXTENSIONS else "IMAGE"
        
        # Generate new name: preset + type + timestamp + index + ext
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        new_name = f"{preset_string}_{file_type}_{timestamp}_{idx}{ext}"
        
        old_path = os.path.join(folder_path, old_name)
        new_path = os.path.join(folder_path, new_name)
        
        os.rename(old_path, new_path)
        if verbose:
            print(f"Renamed: '{old_name}' → '{new_name}'")
    
    if verbose:
        print("Renaming complete!")

# Example Usage
# folder_path = "./media_files/Apple counting"  # Replace with your folder
# preset_string = "Computer_vision_apple_conveyer"     # Customize this

folder_path = "./media_files/People walking"  # Replace with your folder
preset_string = "Computer_vision_object_and_face_detection"     # Customize this
rename_media_files(folder_path, preset_string)
```

## Promt: inside media_files folder, there are multiple folders. folder name also add to preset string
### Deepseek understanding: the enhanced Python function that processes all subfolders inside media_files, incorporating each folder's name into the preset string while categorizing files by type (video/image):
### Keywords:Dynamic Multi-Folder Media File Renamer

```python
import os
from datetime import datetime

def rename_all_media_files(root_folder="media_files", base_preset="Media"):
    """
    Renames all video/image files in all subfolders of root_folder with structure:
    `{base_preset}_{parent_folder}_{file_type}_{timestamp}_{index}.ext`
    
    Supported formats:
    - Videos: .mp4, .mov, .avi, .mkv, .flv, .webm
    - Images: .jpg, .jpeg, .png, .gif, .bmp, .tiff
    
    Args:
        root_folder (str): Parent folder containing subfolders with media files.
        base_preset (str): Base name for all files (default: "Media").
    """
    if not os.path.exists(root_folder):
        print(f"Error: Folder '{root_folder}' does not exist.")
        return

    # Supported extensions
    VIDEO_EXTS = ('.mp4', '.mov', '.avi', '.mkv', '.flv', '.webm')
    IMAGE_EXTS = ('.jpg', '.jpeg', '.png', '.gif', '.bmp', '.tiff')
    
    total_renamed = 0

    # Process each subfolder
    for folder_name in os.listdir(root_folder):
        folder_path = os.path.join(root_folder, folder_name)
        if not os.path.isdir(folder_path):
            continue

        # Slugify folder name
        folder_slug = folder_name.replace(" ", "_").lower()
        
        # Get all media files
        files = [f for f in os.listdir(folder_path) 
                 if f.lower().endswith(VIDEO_EXTS + IMAGE_EXTS)]
        
        if not files:
            print(f"\nNo media files found in '{folder_name}'.")
            continue

        print(f"\nProcessing '{folder_name}' ({len(files)} files):")
        
        for idx, old_name in enumerate(files, 1):
            # Determine file type
            ext = os.path.splitext(old_name)[1].lower()
            file_type = "video" if ext in VIDEO_EXTS else "image"
            
            # Generate new name
            timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
            new_name = f"{base_preset}_{folder_slug}_{file_type}_{timestamp}_{idx}{ext}"
            
            # Rename
            old_path = os.path.join(folder_path, old_name)
            new_path = os.path.join(folder_path, new_name)
            
            os.rename(old_path, new_path)
            print(f"  Renamed [{file_type.upper()}] {old_name} → {new_name}")
            total_renamed += 1

    print(f"\nTotal files renamed: {total_renamed}")

# Example Usage
rename_all_media_files(
    root_folder="./media_files",
    base_preset="computer_vision_object_and_detection_tracking"  # Customize this base name
)
```

## Promt: inside media_files folder, there are multiple folders. folder name also add to preset string except "person" named folder

```python
import os
from datetime import datetime

def rename_media_files_except_person(root_folder="media_files", base_preset="Media"):
    """
    Renames all video/image files in subfolders (except 'person' folders) with structure:
    `{base_preset}_{folder_name}_{file_type}_{timestamp}_{index}.ext`
    
    Args:
        root_folder (str): Parent folder containing subfolders.
        base_preset (str): Base name for all files.
    """
    if not os.path.exists(root_folder):
        print(f"Error: Folder '{root_folder}' does not exist.")
        return

    # Supported extensions
    VIDEO_EXTS = ('.mp4', '.mov', '.avi', '.mkv', '.flv', '.webm')
    IMAGE_EXTS = ('.jpg', '.jpeg', '.png', '.gif', '.bmp', '.tiff')
    
    total_renamed = 0

    for folder_name in os.listdir(root_folder):
        folder_path = os.path.join(root_folder, folder_name)
        
        # Skip if not a folder or folder is named 'person' (case-insensitive)
        if not os.path.isdir(folder_path) or folder_name.lower() == "person":
            continue

        # Slugify folder name
        folder_slug = folder_name.replace(" ", "_").lower()
        
        files = [f for f in os.listdir(folder_path) 
                if f.lower().endswith(VIDEO_EXTS + IMAGE_EXTS)]
        
        if not files:
            print(f"\nNo media files in '{folder_name}' (skipped 'person' folders)")
            continue

        print(f"\nProcessing '{folder_name}':")
        
        for idx, old_name in enumerate(files, 1):
            ext = os.path.splitext(old_name)[1].lower()
            file_type = "video" if ext in VIDEO_EXTS else "image"
            
            timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
            new_name = f"{base_preset}_{folder_slug}_{file_type}_{timestamp}_{idx}{ext}"
            
            old_path = os.path.join(folder_path, old_name)
            new_path = os.path.join(folder_path, new_name)
            
            os.rename(old_path, new_path)
            print(f"  Renamed: {old_name} → {new_name}")
            total_renamed += 1

    print(f"\nTotal renamed: {total_renamed} (skipped 'person' folders)")

# Usage
rename_media_files_except_person(
    root_folder="./media_files",
    base_preset="computer_vision_object_and_detection_tracking"
)
```
